# Player of the Match

```python
df_match_top_batting_performers = df_match.loc[(df_match['result_margin']>0)]
df_player_of_match = df_match_top_batting_performers['player_of_match'].value_counts()
df_player_of_match
```

# Top Batsman as per runs scored

```python
df_batsman_runs = df_ball.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
df_batsman_runs
```

# Total Matches

```python
MatchCountGroupBy = df_ball.groupby(['batsman','id']).count().reset_index()
totalMatchesPlayed = MatchCountGroupBy.groupby(['batsman'])['inning'].count()
totalMatchesPlayed
```

# Average

```python
Average = df_batsman_runs/totalMatchesPlayed
Average.sort_values(ascending=False)
```

# Total Runs as per Over 

```python
totalRunsByAPlayer = df_ball.groupby(['batsman','over']).agg({'batsman_runs':sum})
totalRunsByAPlayer
```

```python
MaxRunsInAnOver = totalRunsByAPlayer['batsman_runs'].groupby('batsman', group_keys=False).nlargest(3).reset_index()
MaxRunsInAnOver
```

```python
MaxRunsInAnOver['position'] = '' 
for i in range(len(MaxRunsInAnOver)):
    if MaxRunsInAnOver['over'].values[i] > 0 and MaxRunsInAnOver['over'].values[i] <= 4: 
         MaxRunsInAnOver['position'].values[i] = 'Opening'
    elif MaxRunsInAnOver['over'].values[i] >= 5 and MaxRunsInAnOver['over'].values[i] <= 8: 
         MaxRunsInAnOver['position'].values[i] = '1th & 2th Down' 
    elif MaxRunsInAnOver['over'].values[i] >= 9 and MaxRunsInAnOver['over'].values[i] <= 12:
         MaxRunsInAnOver['position'].values[i] = '3th & 4th Down' 
    elif MaxRunsInAnOver['over'].values[i] >= 13 and MaxRunsInAnOver['over'].values[i] <= 16:
         MaxRunsInAnOver['position'].values[i] = '5th & 6th Down' 
    elif MaxRunsInAnOver['over'].values[i] >= 17 and MaxRunsInAnOver['over'].values[i] <= 20: 
         MaxRunsInAnOver['position'].values[i] = '7th & 8th Down'
```

# Top Wicket Takers

```python
df_wickets = df_ball.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False)
df_wickets_top_30 = df_wickets.head(30)
df_wickets_top_30
```

```python
print('We also check the number of overs bowled by respective bowlers as below')
```

# Number of runs given by blowers

```python
df_bowler_gave_runs = df_ball.groupby('bowler')['batsman_runs'].sum().sort_values(ascending=True)
df_bowler_gave_runs
```

# Top 30 bowlers on basis of maximum overs

```python
df_overs_of_bowlers = (df_ball['bowler'].value_counts().sort_values(ascending=False))/6
df_overs_of_bowlers_top_30 = df_overs_of_bowlers.head(30)
df_overs_of_bowlers_top_30
```

# Economy of Bowlers

```python
df_bowler_economy = df_bowler_gave_runs/df_overs_of_bowlers_top_30
df_bowler_economy = df_bowler_economy.sort_values(ascending=True)
df_bowler_economy
```

# Wickets to matches Ratio

```python
df_wickets_to_matches_ratio = (df_overs_of_bowlers_top_30/df_wickets_top_30)*df_bowler_economy
df_wickets_to_matches_ratio.sort_values(ascending=True)
```
